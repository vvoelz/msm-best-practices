```python
### IMPORT ###
import pyemma
import os, sys, glob
import matplotlib.pyplot as plt
import numpy as np

# specify data dir
current_dir = './data'

# specify feature used to build msm
# this is mainly done as house-keeping to keep track of what dataset was used
feature_name = 'CA_CA_distances'

# load in pre-generated ITS object from pyemma for plotting
its = pyemma.load(f'{current_dir}/{feature_name}_its.h5')

# make figure
fig, ax = plt.subplots(figsize=(10,7))

# NOTE the 'its' object is in FRAMES (by design)
# pyemma requires specified 'dt' parameter for frame rate in ns
# 100 ps frames for this dataset, so 0.1 ns 
frame_rate_ns = 0.1

# pyemma's own plotting. works like pyplot
pyemma.plots.plot_implied_timescales(its, units='ns', dt=frame_rate_ns, ax=ax)

ax.tick_params(axis='both', which='major', labelsize=20)
ax.tick_params(axis='both', which='minor', labelsize=14)

ax.set_xlabel('lagtime (ns)', fontsize = 20)
ax.set_ylabel('implied timescale (ns)', fontsize = 20)
```
