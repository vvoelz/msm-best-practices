```python
# DEPENDENCEIS
import os, sys, glob
# specific dependencies
import pyemma
import matplotlib.pyplot as plt
import numpy as np

# specify data directory
current_dir = './data'

# filenames of features
features = ['dihedrals','CA_CA_distances','CA_coord']

# make figure
fig, ax = plt.subplots(figsize=(10,7))

# colors of bars
colors = ['red','green','blue']

# location of bar centers on axis
ind = np.arange(len(features))

# for every feature
for f, feature in enumerate(features):
    
    # load in scores from data dir
    data = np.load(f'{current_dir}/{feature}_scores.npy')
    
    # compute average and standard deviation
    avg, std = np.average(data), np.std(data)
    
    # plot scores
    ax.bar(ind[f], avg, yerr=std,
           capsize = 10, label='_nolegend_', color = colors[f])


ax.set_ylabel('VAMP-2 score', size = 20)

ax.set_xticks([0. ,1. ,2.])
ax.tick_params(axis='both', which='major', labelsize=20)
ax.tick_params(axis='both', which='minor', labelsize=14)

#name of axis, may need to change CA to proper Greek/math notation
features_names = ['dihedrals','CA distances','CA coordinates']
ax.set_xticklabels(features_names, rotation = 45)

fig.tight_layout()
plt.show()
```
